```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tab = pd.read_csv('tab_article.csv', sep = ';')
tab
```

```python
fig = plt.figure(figsize=(15,12))
ax1 = fig.add_subplot(111)

x = tab.Year
y1 = tab.Accuracy
y2 = tab.Sensitivity
y3 = tab.Specificity



ax1.scatter(x, y1,color='r', s = 200, marker='^', label='Accuracy', facecolor = 'none')
ax1.scatter(x, y2,color='g', s = 200, marker='*', label='Sensitivity', facecolor = 'none')
ax1.scatter(x, y3,color='b',s = 200, marker='p', label='Specificity', facecolor = 'none')

ax1.set_xlim([2000,2025])
ax1.set_ylim([65,105])
 

#ax1.set_xticks(range(0,21,2))
#On gradue l'axe avec un pas de 2 (au lieu de 2.5)

#ax1.set_xticklabels(["j +" + str(l) for l in range(0,21,2)])
#On modifie les labels sur la graduation

ax1.set_xlabel('Années')

ax1.legend(loc='best')
#permet de choisir l'endroit le plus vide sériesFF
```
