En dessous : fonction définie pour calculer la distance RR, avant de trouver la fonction find_peaks. Elle est donc obsolète

```python
def calculate_RR(row):
    df = pd.DataFrame(row.iloc[:-1])
    df["diff"] = df.iloc[:,0].diff(1)
    df["diff_2"] = -df.iloc[:,0].diff(-1)
    df["flag"] = df.apply(lambda x : 1 if (x["diff"]>0 or pd.isna(x["diff"])) & (x["diff_2"]<=0) & (x.iloc[0]>0.75) 
                          else 0, axis=1)
    try:
        return df[df["flag"] == 1].index[-1] - df[df["flag"] == 1].index[0]
    except:
        return 0
```

### Régression Logistique

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
mit_test = pd.read_csv("../Datascientest/Data/mitbih_test.csv", header=None)
```

```python
X_train = mit_train.iloc[:,:-1]
y_train = mit_train.iloc[:,-1]
X_test = mit_test.iloc[:,:-1]
y_test = mit_test.iloc[:,-1]
```

```python
lr = LogisticRegression()

lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
pd.crosstab(y_test, y_pred, rownames = ["réel"], colnames = ["predict"])
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
